```python
#Here we are creating a function to calculate annualized returns

def summary_ann(returns):
    #Here is the mean daily returns and standard deviation for daily returns
    summary = returns.agg(["mean", "std"]).T
    #Here we are mulitplying our returns by the 252 trading days a year for our total returns
    summary["Return"] = summary["mean"] * 252
    #Here we are calculating risk by multiplying the std dev of returns times the square root of 252 trading days
    summary["Risk"] = summary["std"] * np.sqrt(252)
    summary.drop(columns = ["mean", "std"], inplace = True)
    return summary
```

# Selecting the Tracking Stocks

```python
#Here we are creating a function to help calculate our stock returns

def tracking_ann(returns):
    summary = returns.agg(["mean", "std"]).T
    #Once again we are multiplying by 252 since there are 252 trading days a year
    summary["Premium"] = summary['mean'] * 252
    summary["TE"] = summary['std'] * np.sqrt(252)
    summary.drop(columns = ['mean', 'std'], inplace = True)
    return summary
```

```python
#Here we are creating a new dataframe by applying the function above to ret_diff

tracking = tracking_ann(ret_diff)
```

```python
#Here is the new dataframe
#Premium is return premium
#TE is "tracking error"

tracking.head()
```

```python
tracking.sort_values(by = "TE")
```

```python
#Here we are creating a variable "noa" for "number of assets"
#We want the ten biggest assets

noa = 10
```

```python
#Here we are obtaining the 11 largest assets in the dataframe

tracking_stocks = tracking.sort_values(by='TE').index[:noa+1]
```

# A Simple Portfolio Tracking

```python
#Here we are putting equal weights at the starting point of optimization

eweights = np.full(noa, 1/noa)
```

```python
#Each and every stock has a weight of 10%

eweights
```

```python
#Here we are calculating the daily returns for the equally weighted portfolio

ret2.loc[:, tracking_stocks].mean(axis = 1)
```

```python
#Here we are using the numpy.dot method to calculate returns
#It is the dot product of two arrays

ret2.loc[:, tracking_stocks].dot(eweights)
```

```python
#Here we are creating a function to calculate portfolio returns based on weights

def port_ret(weights):
    return ret2.loc[:, tracking_stocks].dot(weights)
```

```python
#Here we are creating a function to minmize tracking error 

def tracking_error(weights):
    #We are subtracting the DJI returns from our own portfolio return
    #The standard deviation of that difference is the tracking error
    #We annualize the error by multiplying it by the square root of 252
    return port_ret(weights).sub(ret2.DJI).std() * np.sqrt(252)
```

```python
#Here we are applying our tracking error function to the eweights variable

tracking_error(eweights)
```

```python
#Here we are calculating normalized prices
#ew_port = "equally weighted portfolio"

ew_port = port_ret(eweights)[1:].add(1).cumprod().mul(100)
```

```python
ew_port
```

```python
#We are starting at a base value of 100 for our portfolio
#Our equally weighted portfolio is the very last column
#Compare it to the other columns

norm_track["EW_Port"] = ew_port
norm_track.iloc[0,-1] = 100
```

```python
norm_track
```

```python
#Here we are visualizing our equally weighted portfolio vs the DJI

norm_track[["EW_Port", "DJI"]].plot(figsize = (15,10))
plt.legend(fontsize = 20, loc = 2)
plt.show()
```

# The Optimal Tracking Portfolio

```python
#Here we are importing the scipy.optimize library
#SciPy optimize provides functions for minimizing (or maximizing) objective functions, possibly subject to constraints. 
#It includes solvers for nonlinear problems (with support for both local and global optimization algorithms), 
#linear programing, constrained and nonlinear least-squares, root finding, and curve fitting.

import scipy.optimize as sco
```

```python
#We must make sure the weights sum to one

cons = ({'type':'eq', 'fun': lambda x: np.sum(x) - 1})
```

```python
cons
```

```python
#We also need to make sure the bounds of our weights need to be between 0 and 1

bnds = tuple((0,1) for x in range(noa))
```

```python
bnds
```

```python
#Here we are running our first optimization function
#We are minimizing our tracking error
#Method SLSQP uses Sequential Least Squares Programming to minimize a function of several variables 
#with any combination of bounds, equality and inequality constraints.

opts = sco.minimize(tracking_error, eweights, method = "SLSQP", bounds = bnds, constraints = cons)
```

```python
#Here is the output of our optimization
#The first value "fun" is our tracking error annualized 

opts
```

```python
#Here we are calculating the optimal weights for our portfolio

optimal_weights = opts["x"]
```

```python
#Here are the optimal weights

optimal_weights
```

```python
#Here we are creating a panda series attacking the ticker symbols to the optimal weights
#So for JP Morgan (JPM) our optimal weight is 6.3 times e-18

pd.Series(index = tracking_stocks, data = optimal_weights)
```

```python
#Here we are calculating the daily returns of our optimal porfolio

opt_port = port_ret(optimal_weights)[1:].add(1).cumprod().mul(100)
```

```python
opt_port
```

```python
#Here we are adding the optimal porfolio as a column to the main dataframe

norm_track["Opt_Port"] = opt_port
norm_track.iloc[0, -1] = 100
```

```python
#Here is the dataframe in full

norm_track
```

```python
#Here we are visualzing the optimal portfolio vs the DJI returns

norm_track[["Opt_Port", "DJI"]].plot(figsize = (15,10))
plt.legend(fontsize = 20, loc = 2)
plt.show()
```

```python
#Here we are applying our summary function to the different portfolios
#Looks like they all have around the same risk
#The equally weighted portfolio has the greatest return

summary_ann(norm_track[["Opt_Port", "EW_Port", "DJI"]].pct_change())
```
